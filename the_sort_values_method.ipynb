## The sort_values Method
- the _sort_values_ method sorts a __Series__ values in order.
- By default, pandas applies an ascending sort(smallest to largest)
- Customize the sort order with the ascending parameter.

```python
import pandas as pd
```

```python
pokemon = pd.read_csv('Pokemon (2).csv', usecols=['Name'])
google = pd.read_csv('Google_Stock_Price_Train.csv', usecols=['High'])

google.head()
```

```python
google.sort_values(by='High')
```

```python
google.sort_values(by='High', ascending=False)
```

```python
pokemon.head()
```

```python
pokemon.sort_values(by='Name')
pokemon.sort_values(by='Name', ascending=False)
pokemon.sort_values(by='Name', ascending=False).head()
```
